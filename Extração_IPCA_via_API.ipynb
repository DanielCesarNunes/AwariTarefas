#Coleta de dados via API

```python
import pandas as pd
import json, requests
```

```python
codigo_serie = 433
# essa linha pega a resposta do link e atribui as variáveis
response = requests.get(f"http://api.bcb.gov.br/dados/serie/bcdata.sgs.{codigo_serie}/dados?formato=json")
response
```

```python
response.text
```

```python
# realiza a conversão do response para um formato acessível ao python
json_data = json.loads(response.text)
```

```python
json_data
```

```python
ipca = pd.DataFrame(json_data)
```

```python
ipca
```
