```python
# importac√£o das bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
```

```python
# defini√ß√£o da janela desktop
options = webdriver.ChromeOptions()
options.add_argument('--incognito')
options.add_argument('--start-maximized')
options.add_experimental_option('excludeSwitches', ['enable-automation']) 
```

```python
# utilizando o webdriver
driver = webdriver.Chrome(options = options)
```

```python
# carregando a url para a realiza√ß√£o do web-scraping
url = "https://br.linkedin.com/jobs/ci%C3%AAncia-de-dados-vagas?position=1&pageNum=0"
driver.get(url)
time.sleep(3)
```

```python
# localizando os elementos do site
vagas = driver.find_elements_by_class_name('base-card')
```

```python
# iterando para pegar os dados das vagas
lista_empresas = []
lista_locais = []
lista_descricao = []
i = 0

while len(vagas) <= 175:
    for vaga in vagas[i:]:
        vaga.click()
        time.sleep(4)
        try:
            empresa = driver.find_element_by_class_name('topcard__org-name-link')
            local = driver.find_element_by_class_name('topcard__flavor--bullet')
            descricao = driver.find_element_by_class_name('description')
            lista_empresas.append(empresa.text)
            lista_locais.append(local.text)
            lista_descricao.append(descricao.text)
        except:
            pass
    print(f'tamanho lista = {len(lista_empresas)}')
    vagas = driver.find_elements_by_class_name('base-card')
    print(len(vagas))
    i += 25
    print(f'i = {i}')
    if i >= len(vagas):
        break
    
```

```python
df = pd.DataFrame(zip(lista_locais, lista_descricao), index = lista_empresas, columns = ['descri√ß√£o', 'local'])
df
```

```python
for i in range(len(df)):
    print(f'Empresa {i+1}')
    print(f'{df.index[i]}')
    print(f'{df.local[i]}')
    print(f'{df.descri√ß√£o[i]}')
    print()
```

```python
driver.quit()
```
