```python
# importando as bibliotecas necessárias para a realização da tarefa
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# viva real home page

url = 'https://www.vivareal.com.br/aluguel/sp/sao-paulo/zona-sul/cidade-ademar/apartamento_residencial/'

# Utilização do request para realizar a requisição
vivareal_response = requests.get(url)

# Conversão da resposta para HTML através do beautiful soup
vivareal_soup = BeautifulSoup(vivareal_response.text, 'html.parser') 

# Coletando o título da página
vivareal_soup.title
```

```python
# Usando o método prettify para verificar a estrutura html do site
print(vivareal_soup.prettify())
```

```python
# Coleta dos apartamentos
apartamentos = vivareal_soup.find('div', {'class' :'results-main__panel js-list'})
apartamentos
```

```python
# Captura dos elementos necessários para o cálculo do preço do metro quadrado 

lista_enderecos = []
lista_metragem = []
lista_quartos = []
lista_banheiros = []
lista_precos = []

for endereco, metros, quartos, banheiros, precos in zip(apartamentos.find_all('span', {'class':'property-card__address'}),
                                                        apartamentos.find_all('span', {'class':'property-card__detail-value js-property-card-value property-card__detail-area js-property-card-detail-area'}),
                                                        apartamentos.findChildren('li', {'class':'property-card__detail-item property-card__detail-room js-property-detail-rooms'}),
                                                        apartamentos.findChildren('li', {'class':'property-card__detail-item property-card__detail-bathroom js-property-detail-bathroom'}),
                                                        apartamentos.find_all('div', {'class':'property-card__price js-property-card-prices js-property-card__price-small'})):
    lista_enderecos.append(endereco.text)
    lista_metragem.append(metros.text)
    lista_quartos.append(quartos.text)
    lista_banheiros.append(banheiros.text)
    lista_precos.append(precos.text)
  
```

```python
# Retirando os espaços em branco dos dados

for i in range(len(lista_enderecos)):
  lista_enderecos[i] = lista_enderecos[i].strip()
  lista_metragem[i] = lista_metragem[i].strip()
  lista_quartos[i] = lista_quartos[i].strip()
  lista_banheiros[i] = lista_banheiros[i].strip()
  lista_precos[i] = lista_precos[i].strip()

```

```python
# Limpeza de dados
for q in range(len(lista_quartos)):
  lista_quartos[q] = lista_quartos[q][0:2].strip()
  lista_banheiros[q] = lista_banheiros[q][0:2].strip()
  lista_metragem[q] = lista_metragem[q][0:3].strip()
```

```python
# Seperação valor e período do aluguel
for p in range(len(lista_precos)):
  lista_precos[p] = lista_precos[p].replace('R$', ' ').strip()
  lista_precos[p] = lista_precos[p].split(' /')
```

```python
lista_periodo = []
for i in range(len(lista_precos)):
  lista_periodo.append(lista_precos[i][1][0:3])
  lista_precos[i] = lista_precos[i][0]

for j in range(len(lista_precos)):
  lista_precos[j] = lista_precos[j].replace('.', '')
```

```python
lista_precos
```

```python
lista_logradouro = []
lista_cidade = []
lista_estado = []

for i in range(len(lista_enderecos)):
  lista_enderecos[i] = lista_enderecos[i].split('-')

for endereco in lista_enderecos:
  if len(endereco) == 3:
    lista_logradouro.append(endereco[0].strip())
    lista_cidade.append(endereco[1].strip())
    lista_estado.append(endereco[2].strip())
  else:
    lista_logradouro.append(' ')
    lista_cidade.append(endereco[0].strip())
    lista_estado.append(endereco[1].strip())
```

```python
lista_cidade
```

```python
for i in range(len(lista_cidade)):
  lista_cidade[i] = lista_cidade[i].split(',')
```

```python
lista_bairro = []
lista_sp = []
for bairro in lista_cidade:
  lista_bairro.append(bairro[0].strip())
  lista_sp.append(bairro[1].strip())
```

```python
# Colocando os dados obtidos em um DataFrame
imoveis = pd.DataFrame({'Cidade': lista_sp, 'Bairro':lista_bairro, 'Endereco': lista_logradouro,  'Metros':lista_metragem,
                        'Quartos': lista_quartos, 'Banheiros': lista_banheiros, 'Valor Aluguel R$': lista_precos})
imoveis.head(10)
```

```python
imoveis.loc[imoveis.Quartos == '--','Quartos']='0'
imoveis.loc[imoveis.Banheiros == '--','Banheiros']='0'
imoveis.head(10)
```

```python
# Verificação dos tipos dos dados
imoveis.dtypes
```

```python
# alteração dos tipos de dados do dataframe
imoveis[['Metros', 'Quartos', 'Banheiros', 'Valor Aluguel R$']] = imoveis[['Metros', 'Quartos',
                                                                           'Banheiros','Valor Aluguel R$']].apply(pd.to_numeric)

                                                                      
```

```python
imoveis.dtypes
```

```python
# Cálculo da média do valor do metro quadrado por bairro
media_aluguel = round(imoveis.groupby('Bairro')[['Metros', 'Valor Aluguel R$']].mean(), 2)
media_aluguel
```

```python
# Valor do m²
valor = imoveis['Valor Aluguel R$'].mean() / imoveis['Metros'].mean()
valor
```

```python
# Calculadora do valor do aluguel
metros = 0

while metros <= 0:
  metros = float(input("Digite o tamanho do apartamento (m²): "))
  if metros <= 0:
    print('Valor inválido. Digite novamente!')

aluguel = round(metros * valor, 2)

print(f'Bairro: Cidade Ademar\nMetros: {metros}\nAluguel R$ {aluguel}')
```
